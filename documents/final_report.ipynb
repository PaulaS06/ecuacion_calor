Por otro lado, la ley de **Calorimetr√≠a** o **ecuaci√≥n del calor sensible**, es una ley f√≠sica que describe la cantidad de calor ($\Delta Q$) que un cuerpo gana o pierde al cambiar su temperatura sin cambio de estado.

Para obtener la masa de la barra se puede usar la densidad ($\rho$) de la barra, que es la masa por unidad de volumen.
La densidad se puede expresar como:
$$ \rho = \frac{m}{V} $$

Seguiendo esto, la masa se puede expresar como:
$$ m = \rho \cdot V $$

Y sabemos que el volumen es: A * $\Delta$ x, entonces: 

$$ m = \rho \cdot A \cdot \Delta x $$

Si reemplazamos en la ecuaci√≥n de calorimetr√≠a, se obtiene que:
$$ \Delta Q = \rho \cdot A \cdot \Delta x \cdot c \cdot \Delta u $$


Esta ecuaci√≥n representa la energ√≠a t√©rmica en el espacio, sin embargo se necesita la energ√≠a t√©rmica por unidad de tiempo, por lo que se debe dividir entre el tiempo ($\Delta t$), as√≠: 

$$ \frac{\Delta Q}{\Delta t} = \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\Delta u}{\Delta t} $$

Se necesita que $\Delta x$ y $\Delta t$ tiendan a cero, para que la ecuaci√≥n sea continua, entonces se puede expresar la derivada como:

$$ \frac{\partial Q}{\partial t} = \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} $$

En conclusi√≥n se obtuvo el cambio de la energ√≠a t√©rmica por cada intervalo de tiempo. Lo que equivale a la energ√≠a que entra menos la energ√≠a que sale m√°s la energ√≠a generada. Reemplazando en la ecuaci√≥n se obtiene:

$$ \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} = A \cdot \Phi(x, t) + A \cdot \Phi(x + \Delta x, t) + A \cdot \Delta x \cdot E(x, t)$$

En esta ecuaci√≥n se puede cancelar el √°rea de la secci√≥n transversal de la barra, ya que aparece en ambos lados de la ecuaci√≥n como una constante. Es impotante aclarar que esto se puede realizar ya que se est√° trabajando con la ecuaci√≥n de calor en una dimensi√≥n.

$$ \rho \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} = \Phi(x, t) + \Phi(x + \Delta x, t) + \Delta x \cdot E(x, t)$$

Se debe aplicar el l√≠mite cuando Œîùë• ‚Üí 0. Este proceso implica el uso de derivadas e integraci√≥n para transformar la ecuaci√≥n en su forma continua, as√≠: 

1. Se toma un intervalo arbitrario de [a, b] para integrar la ecuaci√≥n.

La integral de la ecuaci√≥n del lado derecho se puede expresar como:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx $$

Para integrar el lado izquierdo es necesario multiplicar y dividir los dos primeros terminos por $\Delta x$, as√≠: 

$$\frac{\Phi(x, t) + \Phi(x + \Delta x, t)}{\Delta x} \Delta x + \Delta x \cdot E(x, t)$$

Aplicando el l√≠mite cuando Œîùë• ‚Üí 0, se obtiene que el primer t√©rmino se convierte en la derivada de la funci√≥n $\Phi(x, t)$ y el segundo t√©rmino se convierte en la funci√≥n $E(x, t)$, as√≠: 

$$ -\frac{\partial \Phi}{\partial x} \Delta x + E(x, t)$$

Si integramos el lado izquiero e igualamos las ecuaciones integradas se obtiene que:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx = - \int_a^b \frac{\partial \Phi}{\partial x} dx + \int_a^b E(x, t) dx$$

Como se tienen integrales definidas con los mismos l√≠mites, se puede aplicar el teorema fundamental del c√°lculo. Adem√°s podemos igualar todo a cero, as√≠:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx + \int_a^b \frac{\partial \Phi}{\partial x} dx - \int_a^b E(x, t) dx = 0$$

$$
\int_a^b \left[ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial \Phi}{\partial x} - E(x, t) \right] dx = 0
$$


Como [a,b] es un intervalo arbitrario, se puede concluir que el integrando debe ser cero para cualquier intervalo. Para esto se puede aplicar el teorema del calculo. 
**Teorema del calculo**: Si f es continua en un intervalo cerrado [a,b] y la integral de f en [a,b] es cero, entonces f(x) = 0 para todo x en [a,b]. Es decir: 

$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial \Phi}{\partial x} - E(x, t) = 0$$

En este punto podemos utilizar la ley de Fourier para expresar el flujo de energ√≠a como:

$$ \Phi(x, t) = -k \cdot \frac{\partial u}{\partial x} $$
Donde k es la conductividad t√©rmica del material y k > 0.

La funci√≥n $\Phi(x, t)$ cumple que: 
1. Si la temperatura es constante, el flujo de energ√≠a es cero.
2. Si hay diferencia de temperatura, entonces la energ√≠a t√©rmica fluye de la regi√≥n caliente(alta temperatura) a la regi√≥n fr√≠a(baja temperatura).
3. A mayor diferencia de temperatura, mayor flujo de energ√≠a.

Seg√∫n esto, podemos expresar la igualdad como:
$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial}{\partial x}\left(-k \cdot \frac{\partial u}{\partial x}\right) - E(x, t) = 0$$ 
$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} - k \cdot \frac{\partial^2 u}{\partial x^2} - E(x, t) = 0$$

Para esta ecuaci√≥n se puede asumir que la densidad y el calor espec√≠fico son constantes, as√≠ como la conductividad t√©rmica. Por lo que se puede expresar como:

$$ \frac{\partial u}{\partial t} - \alpha \cdot \frac{\partial^2 u}{\partial x^2} - E(x, t) = 0$$

Donde 
$$ \alpha = \frac{k}{\rho \cdot c} $$

En este caso el √∫ltimo t√©rmino de la ecuaci√≥n se puede despreciar, ya que este existe y se debe calcular cuando se ha ganado energ√≠a t√©rmica en el intervalo, mientras no exista no es necesario tenerlo. Por lo que se puede expresar como:

$$ \frac{\partial u}{\partial t} = \alpha \cdot \frac{\partial^2 u}{\partial x^2}$$

De esta manera se puede llegar a la ecuaci√≥n diferencial parcial de calor en una dimensi√≥n.

# M√©todo de Diferencias Finitas

El m√©todo de diferencias finitas es un m√©todo num√©rico que se usa para resolver ecuaciones diferenciales. Consiste en aproximar las derivadas de la ecuaci√≥n diferencial con expresiones algebraicas. 

Seg√∫n este m√©todo, la derivada parcial de primer orden se aproxima utilizando la siguiente expresi√≥n algebraica:

En este caso para hacerlo con respecto a la variable _x_:

$$ \frac{\partial u}{\partial t} \approx \frac {u_{i+1} - u_i}{\Delta x} $$
$$ \frac{\partial u}{\partial t} \approx \frac {u_{i+1}^n- u_i^n}{\Delta x} $$

Y para hacerlo con respecto a la variable _t_:

$$ \frac{\partial u}{\partial t} \approx \frac {u^{n+1} - u^n}{\Delta t} $$
$$ \frac{\partial u}{\partial t} \approx \frac {u_i^{n+1} - u_i^n}{\Delta t} $$


Y la derivada parcial de segundo orden se aproxima utilizando la siguiente expresi√≥n algebraica: 
En este caso para hacerlo con respecto a la variable _x_:

$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_{i-1} - 2u_i + u_{i+1} \right)}{\Delta t^2}
$$
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_{i-1}^n - 2u_i^n + u_{i+1}^n \right)}{\Delta t^2}
$$

Y para hacerlo con respecto a la variable _t_:
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u^{n-1} - 2u^n + u^{n+1} \right)}{\Delta t^2}
$$
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_i^{n-1} - 2u_i^n + u_i^{n+1} \right)}{\Delta t^2}
$$

Estas son formulas que salieron de la aproximaci√≥n de Taylor, que es una forma de aproximar funciones usando polinomios.

Teniendo clara la formula de diferencias finitas que se puede utilizar, teniendo la ecuaci√≥n de calor: 
$$ \frac{\partial u}{\partial t} = \alpha \cdot \frac{\partial^2 u}{\partial x^2}$$

Se puede reemplazar la ecuaci√≥n de calor en la ecuaci√≥n de diferencias finitas, as√≠:
$$ \frac{u_i^{n+1} - u_i^n}{\Delta t} = \alpha \cdot \frac{u_{i-1}^n - 2u_i^n + u_{i+1}^n}{\Delta x^2} $$

Para determinar la ecuaci√≥n que calcula la temperatura del siguiente paso, tenemos que despejar $u_i^{n+1}$, as√≠: 

$$ u_i^{n+1} - u_i^n = \alpha \cdot \Delta t \cdot \frac{u_{i-1}^n - 2u_i^n + u_{i+1}^n}{\Delta x^2} $$
$$ u_i^{n+1} - u_i^n = \alpha \cdot \frac{\Delta t}{\Delta x^2} \cdot u_{i-1}^n - 2u_i^n + u_{i+1}^n $$
$$ u_i^{n+1} = \left[\alpha \cdot \frac{\Delta t}{\Delta x^2} \cdot \left(u_{i-1}^n - 2u_i^n + u_{i+1}^n\right) \right] +  u_i^n$$
