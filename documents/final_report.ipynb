# Fundamento Teórico - Ecuación de calor 
La ecuación de calor es una ecuación diferencial en derivadas parciales lineales de segundo grado. 
Para comprender esto es necesario saber que una **derivada parcial** es la derivada de una función con respecto a una de sus variables, manteniendo las demás constantes. Estas son usadas para entender cómo cambia una función en relación con cada variable independiente y analizar cómo afecta cada una en el resultado de la función. 

Para representar una derivada parcial se utiliza el símbolo ∂ y un ejemplo claro de este concepto es: 


- Si se tiene una función es _z = f(x, y)_: 

Las derivadas parciales son la derivada de _z_ con respecto a _x_ considerando _y_ como constante. 
$$
\frac{\partial z}{\partial x} = \frac{\partial f(x,y)}{\partial x}
$$

Y la derivada de _z_ con respecto a _y_ considerando _x_ como constante. 
$$
\frac{\partial z}{\partial y} = \frac{\partial f(x,y)}{\partial y}
$$

Normalmente las ecuaciones diferenciales parciales son fundamentales en la comprensión científica moderna del sonido, el calor, la difusión, la electrostática, el electrodinámica, la termodinámica, la dinámica de fluidos, la elasticidad, la relatividad general y la mecánica cuántica. 

En este proyecto se trabajará sobre la ecuación de calor, inicialmente en una dimensión. 


## Ecuación de calor en una dimensión

La ecuación de calor se deriva del principio de conservación de la energía y la ley de Fourier del flujo de calor. Establece que la velocidad de cambio de la temperatura en un punto es proporcional a la segunda derivada parcial de la temperatura, lo cual implica que el calor se distribuye desde las zonas de mayor temperatura hacia las de menor temperatura, así: 

$$
\frac{∂u}{∂t}= α\frac{∂^2 u}{∂x^2 }
$$

# Fundamento Físico - Ecuación de calor en una dimensión

La ecuación de calor proviene del planteamiento del cambio de temperatura en un punto de una barra unidimensional.

En este punto se realizará la explicación detallada de donde sale la ecuación de calor.
$$
\frac{∂u}{∂t}= α\frac{∂^2 u}{∂x^2 }
$$

Para entender el origen de esta ecuación, se va a tomar una barra unidimensional, de la cual vamos a tomar una parte para analizarla.

<center>
<img src="anexo\barra.png" alt="Barra unidimensional" width="300" height="100"/>
</center>

Inicialmente se tiene claro que: 
- La temperatura es constante en cualquier parte de una sección transversal de la barra.
- La densidad, masa y área de la barra son constantes en cualquier parte de la barra.
- La superficie lateral de la barra es aislante, por lo que no hay pérdida de calor.


**Flujo de energía**: cantidad de energía térmica por unidad tiempo y unidad de superficie. Esta se representa con  &Phi;. 
- positiva -> a la derecha (&Phi; > 0)
- negativa -> a la izquierda (&Phi; < 0)

**Energía térmica generada en el interior**: esto se puede dar por reacciones químicas, corriente elétrica, etc. Esta es la energía generada por unidad de volumen y unidad de tiempo. Se representa como _E_.

_E_ = _E_ (x, t)

<br><br>
Según esto el flujo de energía que: 
- ENTRA: &Phi;(x, t)
- SALE: &Phi;(x + $\Delta$ x, t)

`Energía térmica por unidad de tiempo = Energía que entra - Energía que sale + Energía generada`

Para poner la formula en términos matemáticos se tiene que:
- La energía que entra y que sale se puede expresar como el **flujo de energía multiplicado por el área de la sección transversal de la barra**. Esto se da porque el flujo de energía es la cantidad de energía térmica por unidad de tiempo y unidad de superficie.
- La energía generada se puede expresar como **la energía generada multiplicada por el volumen de la barra**. Esto se da porque la energía que se genera adentro de la barra es la energía generada por unidad de volumen y unidad de tiempo.

El volumen de la barra es el área de la sección transversal multiplicada por el alto de la barra.
volumen = área * alto = A * $\Delta$ x

Energía térmica por unidad de tiempo =
$A \cdot \Phi(x, t) + A \cdot \Phi(x + \Delta x, t) + V \cdot E(x, t)$

Reemplazando el volumen: 

<center>

**Energía térmica por unidad de tiempo** = 
$A \cdot \Phi(x, t) + A \cdot \Phi(x + \Delta x, t) + A \cdot \Delta x \cdot E(x, t)$

</center>



Por otro lado, la ley de **Calorimetría** o **ecuación del calor sensible**, es una ley física que describe la cantidad de calor ($\Delta Q$) que un cuerpo gana o pierde al cambiar su temperatura sin cambio de estado.
Esta se puede expresar como:
$\Delta Q = m \cdot c \cdot \Delta T$ 

Donde:
m = masa del cuerpo <br>
c = calor específico del cuerpo <br>
$\Delta T$ = cambio de temperatura <br>
$\Delta Q$ = energía termica -> en este contexto se nombrará  $\Delta u$<br>

<center>

$\Delta Q = m \cdot c \cdot \Delta u$ 

</center>

Para obtener la masa de la barra se puede usar la densidad ($\rho$) de la barra, que es la masa por unidad de volumen.
La densidad se puede expresar como:
$$ \rho = \frac{m}{V} $$

Seguiendo esto, la masa se puede expresar como:
$$ m = \rho \cdot V $$

Y sabemos que el volumen es: A * $\Delta$ x, entonces: 

$$ m = \rho \cdot A \cdot \Delta x $$

Si reemplazamos en la ecuación de calorimetría, se obtiene que:
$$ \Delta Q = \rho \cdot A \cdot \Delta x \cdot c \cdot \Delta u $$


Esta ecuación representa la energía térmica en el espacio, sin embargo se necesita la energía térmica por unidad de tiempo, por lo que se debe dividir entre el tiempo ($\Delta t$), así: 

$$ \frac{\Delta Q}{\Delta t} = \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\Delta u}{\Delta t} $$

Se necesita que $\Delta x$ y $\Delta t$ tiendan a cero, para que la ecuación sea continua, entonces se puede expresar la derivada como:

$$ \frac{\partial Q}{\partial t} = \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} $$

En conclusión se obtuvo el cambio de la energía térmica por cada intervalo de tiempo. Lo que equivale a la energía que entra menos la energía que sale más la energía generada. Reemplazando en la ecuación se obtiene:

$$ \rho \cdot A \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} = A \cdot \Phi(x, t) + A \cdot \Phi(x + \Delta x, t) + A \cdot \Delta x \cdot E(x, t)$$

En esta ecuación se puede cancelar el área de la sección transversal de la barra, ya que aparece en ambos lados de la ecuación como una constante. Es impotante aclarar que esto se puede realizar ya que se está trabajando con la ecuación de calor en una dimensión.

$$ \rho \cdot \Delta x \cdot c \cdot \frac{\partial u}{\partial t} = \Phi(x, t) + \Phi(x + \Delta x, t) + \Delta x \cdot E(x, t)$$

Se debe aplicar el límite cuando Δ𝑥 → 0. Este proceso implica el uso de derivadas e integración para transformar la ecuación en su forma continua, así: 

1. Se toma un intervalo arbitrario de [a, b] para integrar la ecuación.

La integral de la ecuación del lado derecho se puede expresar como:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx $$

Para integrar el lado izquierdo es necesario multiplicar y dividir los dos primeros terminos por $\Delta x$, así: 

$$\frac{\Phi(x, t) + \Phi(x + \Delta x, t)}{\Delta x} \Delta x + \Delta x \cdot E(x, t)$$

Aplicando el límite cuando Δ𝑥 → 0, se obtiene que el primer término se convierte en la derivada de la función $\Phi(x, t)$ y el segundo término se convierte en la función $E(x, t)$, así: 

$$ -\frac{\partial \Phi}{\partial x} \Delta x + E(x, t)$$

Si integramos el lado izquiero e igualamos las ecuaciones integradas se obtiene que:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx = - \int_a^b \frac{\partial \Phi}{\partial x} dx + \int_a^b E(x, t) dx$$

Como se tienen integrales definidas con los mismos límites, se puede aplicar el teorema fundamental del cálculo. Además podemos igualar todo a cero, así:

$$ \int_a^b \rho \cdot c \cdot \frac{\partial u}{\partial t} dx + \int_a^b \frac{\partial \Phi}{\partial x} dx - \int_a^b E(x, t) dx = 0$$

$$
\int_a^b \left[ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial \Phi}{\partial x} - E(x, t) \right] dx = 0
$$


Como [a,b] es un intervalo arbitrario, se puede concluir que el integrando debe ser cero para cualquier intervalo. Para esto se puede aplicar el teorema del calculo. 
**Teorema del calculo**: Si f es continua en un intervalo cerrado [a,b] y la integral de f en [a,b] es cero, entonces f(x) = 0 para todo x en [a,b]. Es decir: 

$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial \Phi}{\partial x} - E(x, t) = 0$$

En este punto podemos utilizar la ley de Fourier para expresar el flujo de energía como:

$$ \Phi(x, t) = -k \cdot \frac{\partial u}{\partial x} $$
Donde k es la conductividad térmica del material y k > 0.

La función $\Phi(x, t)$ cumple que: 
1. Si la temperatura es constante, el flujo de energía es cero.
2. Si hay diferencia de temperatura, entonces la energía térmica fluye de la región caliente(alta temperatura) a la región fría(baja temperatura).
3. A mayor diferencia de temperatura, mayor flujo de energía.

Según esto, podemos expresar la igualdad como:
$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} + \frac{\partial}{\partial x}\left(-k \cdot \frac{\partial u}{\partial x}\right) - E(x, t) = 0$$ 
$$ \rho \cdot c \cdot \frac{\partial u}{\partial t} - k \cdot \frac{\partial^2 u}{\partial x^2} - E(x, t) = 0$$

Para esta ecuación se puede asumir que la densidad y el calor específico son constantes, así como la conductividad térmica. Por lo que se puede expresar como:

$$ \frac{\partial u}{\partial t} - \alpha \cdot \frac{\partial^2 u}{\partial x^2} - E(x, t) = 0$$

Donde 
$$ \alpha = \frac{k}{\rho \cdot c} $$

En este caso el último término de la ecuación se puede despreciar, ya que este existe y se debe calcular cuando se ha ganado energía térmica en el intervalo, mientras no exista no es necesario tenerlo. Por lo que se puede expresar como:

$$ \frac{\partial u}{\partial t} = \alpha \cdot \frac{\partial^2 u}{\partial x^2}$$

De esta manera se puede llegar a la ecuación diferencial parcial de calor en una dimensión.

# Discretización 

En este punto se convertirá la ecuación de calor, que es una dimension que trabaja con variables continuas (como espacio y tiempo) en una versión discreta que pueda resolverse por medio de la computación. 

Una opción para hacer esto es usar el método de **diferencias finitas**, que consiste en aproximar las derivadas de la ecuación por diferencias entre valores discretos de la función.

Un ejemplo común del uso de las diferencias finitas es el método de _Euler_, usado en la computación y resolución de problemas físicos. 

En este proyecto se utilizará el método de diferencias finitas para resolver la ecuación de calor en una barra unidimensional.

Para la discretización es necesario definir el espacio y el tiempo en el que se va a trabajar.

- **Espacio:** Se define el espacio como una malla unidimensional, donde cada punto de la malla representa un punto en la barra. La distancia entre dos puntos consecutivos es $\Delta x$ y el número de puntos en la barra es _N_.

    El instante en cada punto de la barra se puede expresar como:
    $$ x_i = i \cdot \Delta x $$
    donde i es el índice del punto en la barra y $\Delta x$ es la distancia entre dos puntos consecutivos.

- **Tiempo:** Se define el tiempo como una malla unidimensional, donde cada punto de la malla representa un instante en el tiempo. La distancia entre dos puntos consecutivos es $\Delta t$ y el número de puntos en el tiempo es _M_.

    El instante en cada punto de la barra se puede expresar como:
    $$t^n = n \cdot \Delta t$$
    donde n es el índice del instante en el tiempo y $\Delta t$ es la distancia entre dos puntos consecutivos.

# Método de Diferencias Finitas

El método de diferencias finitas es un método numérico que se usa para resolver ecuaciones diferenciales. Consiste en aproximar las derivadas de la ecuación diferencial con expresiones algebraicas. 

Según este método, la derivada parcial de primer orden se aproxima utilizando la siguiente expresión algebraica:

En este caso para hacerlo con respecto a la variable _x_:

$$ \frac{\partial u}{\partial t} \approx \frac {u_{i+1} - u_i}{\Delta x} $$
$$ \frac{\partial u}{\partial t} \approx \frac {u_{i+1}^n- u_i^n}{\Delta x} $$

Y para hacerlo con respecto a la variable _t_:

$$ \frac{\partial u}{\partial t} \approx \frac {u^{n+1} - u^n}{\Delta t} $$
$$ \frac{\partial u}{\partial t} \approx \frac {u_i^{n+1} - u_i^n}{\Delta t} $$


Y la derivada parcial de segundo orden se aproxima utilizando la siguiente expresión algebraica: 
En este caso para hacerlo con respecto a la variable _x_:

$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_{i-1} - 2u_i + u_{i+1} \right)}{\Delta t^2}
$$
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_{i-1}^n - 2u_i^n + u_{i+1}^n \right)}{\Delta t^2}
$$

Y para hacerlo con respecto a la variable _t_:
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u^{n-1} - 2u^n + u^{n+1} \right)}{\Delta t^2}
$$
$$
\frac{\partial^2 u}{\partial t^2} \approx \frac {\left( u_i^{n-1} - 2u_i^n + u_i^{n+1} \right)}{\Delta t^2}
$$

Estas son formulas que salieron de la aproximación de Taylor, que es una forma de aproximar funciones usando polinomios.

Teniendo clara la formula de diferencias finitas que se puede utilizar, teniendo la ecuación de calor: 
$$ \frac{\partial u}{\partial t} = \alpha \cdot \frac{\partial^2 u}{\partial x^2}$$

Se puede reemplazar la ecuación de calor en la ecuación de diferencias finitas, así:
$$ \frac{u_i^{n+1} - u_i^n}{\Delta t} = \alpha \cdot \frac{u_{i-1}^n - 2u_i^n + u_{i+1}^n}{\Delta x^2} $$

Para determinar la ecuación que calcula la temperatura del siguiente paso, tenemos que despejar $u_i^{n+1}$, así: 

$$ u_i^{n+1} - u_i^n = \alpha \cdot \Delta t \cdot \frac{u_{i-1}^n - 2u_i^n + u_{i+1}^n}{\Delta x^2} $$
$$ u_i^{n+1} - u_i^n = \alpha \cdot \frac{\Delta t}{\Delta x^2} \cdot u_{i-1}^n - 2u_i^n + u_{i+1}^n $$
$$ u_i^{n+1} = \left[\alpha \cdot \frac{\Delta t}{\Delta x^2} \cdot \left(u_{i-1}^n - 2u_i^n + u_{i+1}^n\right) \right] +  u_i^n$$
